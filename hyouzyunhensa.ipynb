```python
#グローバル変数
cir = 's5378'  # 対象回路
outval_file_sum = 6  # 変換対象の故障辞書ファイルの数

delited_data_num_file = cir + '分割正解データ削除後/' + cir + 'delited_data_num'  # 削除後の正解データ数ファイル


def main():
    with open(delited_data_num_file, 'r') as f:
        line = f.readline().strip()
        average_delited_data_num = float(line.split(":")[1])  # 削除後の正解データ数
        lines = f.readlines()[3:]
        delited_data_num = [int(s.strip().split(":")[1]) for s in lines]

        print("削除後の正解データ数:", average_delited_data_num)
        print(delited_data_num)
        print(len(delited_data_num))
    
    variance = 0.0
    for i in range(len(delited_data_num)):
        variance += (delited_data_num[i] - average_delited_data_num) ** 2

    hyouzyunhensa = (variance / len(delited_data_num)) ** 0.5
    print("削除後の正解データの標準偏差:", hyouzyunhensa)
        


if __name__ == "__main__":
    main()
```
