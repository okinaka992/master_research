```python

#グローバル変数
cir = 's38584'  # 対象回路
model_folder = cir + 'sepmodel2/'  # 全てのモデルを学習させる場合に、学習済みモデルを保存するフォルダ

finished_model_file = 'finished_model.txt'
not_100_percent_models_file = model_folder + cir + 'not_100_percent_models'  # 初回からre_ann_learn2.py, re_ann_learn3.pyで学習できるようにnot_100_percent_models_fileにすべてのモデル番号を保存する

with open(finished_model_file, "r") as f:
    lines = f.readlines()
    model_ID = [int((line.split()[0]).split("：")[1]) for line in lines]
    acc = [(line.split()[1]).split("：")[1] for line in lines]
    
with open(not_100_percent_models_file, "r") as f:
    lines = f.readlines()
    top = lines[0]
    not_100_models = [int(line.strip()) for line in lines[1:]]

model_num = len(not_100_models)
print(model_num)

model_ID_sorted, acc_sorted = zip(*sorted(zip(model_ID, acc)))
    
missing = [x for x in not_100_models if x not in model_ID_sorted]
print(missing)

a = [x for x in acc if x != "1.0"]
print(len(a))

with open("a.txt", "w") as f:
    for a in acc_sorted:
        f.write(a + "\n")

acc_sorted = list(acc_sorted)
for i in missing:
    acc_sorted.insert(i, "0.0")

with open("aa.txt", "w") as f:
    for a in acc_sorted:
        f.write(a + "\n")
        
print(len(acc))

a = [x for x in acc_sorted if x != "1.0"]
print(len(a))

with open(not_100_percent_models_file, "w") as f:
    f.write(top)
    for i in range(model_num):
        if acc_sorted[i] != "1.0":
            f.write(str(i) + "\n")
            
```
