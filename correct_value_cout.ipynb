```python
# 正解データ中の値の種類の総数を求めるプログラム
# cd workspace/research2/experiment
#　実行コマンド：　python3 learn_data_suplit.py

#グローバル変数
cir = 's5378'  # 対象回路
suplit_num = None # 何個ずつ分割するか
correct_data_folder = cir + '分割正解データ/' # 正解データを分割して保存するフォルダ
suplited_model_num_file = correct_data_folder + cir + 'suplit_num'  # 分割されたモデルの数を保存するファイル
delited_correct_data_folder = cir + '分割正解データ削除後/'  # 削除後の分割正解データを保存するフォルダ
suplit_correct_data_file = delited_correct_data_folder + cir + 'integrated_output'  # 分割された正解データファイル
correct_data_value_file = correct_data_folder + cir + 'correct_value'  # 正解データの種類（値）を保存するファイル
suplit_num_file = cir + 'suplit_num'  # 分割数を保存するファイル

correct_value_count_file = delited_correct_data_folder + "/" + cir + 'correct_value_count'  # 正解データの値の数を保存するファイル
correct_value_count_sorted_file = delited_correct_data_folder + "/" + cir + 'correct_value_count_sorted'  # 正解データの値の数を降順に保存するファイル
correct_middle_value_count_file = delited_correct_data_folder + "/" + cir + 'correct_middle_value_count'  # 正解データの中間値の数を保存するファイル
correct_middle_value_count_sorted_file = delited_correct_data_folder + "/" + cir + 'correct_middle_value_count_sorted'  # 正解データの中間値の数を降順に保存するファイル
correct_data_num_sorted_file = delited_correct_data_folder + "/" + cir + 'delited_data_num_sorted'  # 正解データの総数を降順に保存するファイル

correct_value = None  # 正解データの値
suplited_model_num = None  # 分割されたモデルの数


with open(suplited_model_num_file, 'r') as f:
    suplited_model_num = int(f.readline().strip())


with open(correct_data_value_file, 'r') as f:
    correct_value = [float(x) for x in f.readline().strip().split()]

print("correct_value:", correct_value)


sum_correct_value_count = [[0 for _ in range(len(correct_value))] for _ in range(suplited_model_num)]  # 正解データの値のカウント用リスト
sum_correct_data_count = [0] * suplited_model_num  # 各モデルの正解データの総数を格納するリスト
for i in range(suplited_model_num):
    correct_data = []
    with open(suplit_correct_data_file + str(i), 'r') as f:
        lines = [_.replace("\n", "") for _ in f.readlines()]  # 各行の空白文字と改行文字を削除
        correct_data = [list(_.split(',')) for _ in lines]  # 1次元リストを2次元リストに変換
    
    sum_correct_data_count[i] = len(correct_data) # 各モデルの正解データの総数を計算
    for j in range(len(correct_data)):
        for k in range(len(correct_data[j])):
            value = float(correct_data[j][k])
            if value in correct_value:
                idx = correct_value.index(value)
                sum_correct_value_count[i][idx] += 1

print("correct_data[0]:", correct_data[0])
print("sum_correct_value_count:", sum_correct_value_count)


# 正解データの総数を降順に並べる
sorted_by_data_count = sorted(enumerate(sum_correct_data_count, start=0), key=lambda x: x[1], reverse=True)
with open(correct_data_num_sorted_file, 'w') as f:
    for model_id, total in sorted_by_data_count:
        f.write(f"正解データ{model_id} : {total}\n")


# 各モデルにおける0.25, 0.75の数の合計を計算
sum_middle_value_count = [0] * suplited_model_num  # 0.25と0.75の数を格納するリスト
for i in range(suplited_model_num):
    sum_middle_value_count[i] += sum_correct_value_count[i][1] + sum_correct_value_count[i][2] 

print("sum_middle_value_count:", sum_middle_value_count)


with open(correct_value_count_file, 'w') as f:
    f.write(str(correct_value) + '\n')
    for i in range(suplited_model_num):
        f.write("モデル" + str(i+1) + " の正解データの値の数:\n")
        f.write(str(sum_correct_value_count[i][0]) + ', ' + str(sum_correct_value_count[i][1]) + ', ' + str(sum_correct_value_count[i][2]) + ', ' + str(sum_correct_value_count[i][3]) + '\n')


with open(correct_middle_value_count_file, 'w') as f:
    for i in range(suplited_model_num):
        f.write("モデル" + str(i+1) + " の正解データの中間値の数:\n")
        f.write(str(sum_middle_value_count[i]) + '\n')


# 各モデルの総数を計算
sum_correct_data = [sum(row) for row in sum_correct_value_count]
print("sum_correct_data:", sum_correct_data)

# モデル番号付きで総数を降順に並べる（モデル番号は0始まり）
sorted_by_total = sorted(enumerate(sum_correct_data, start=0), key=lambda x: x[1], reverse=True)
with open(correct_value_count_sorted_file, 'w') as f:
    f.write(str(correct_value) + '\n')
    f.write("モデル番号, 総数\n")
    for model_id, total in sorted_by_total:
        f.write(f"モデル{model_id} : {total}\n")

# 中間値の数をモデル番号付きで降順に並べる
sorted_by_middle = sorted(enumerate(sum_middle_value_count, start=0), key=lambda x: x[1], reverse=True)
with open(correct_middle_value_count_sorted_file, 'w') as f:
    f.write("モデル番号, 中間値の数\n")
    for model_id, total in sorted_by_middle:
        f.write(f"モデル{model_id} : {total}\n")
```
