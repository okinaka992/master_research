```python
# モデルの総メモリ量を計算するプログラム

#グローバル変数
cir = 's38584'  # 対象回路
model_folder = cir + 'sepmodel2/'  # 分割モデルフォルダ
input_data_file = cir + '分割入力データ/' + cir + 'input'  # 入力データフォルダ
suplited_correct_data_folder = cir + '分割正解データ/'  # 削除後の正解データフォルダ
suplited_delited_correct_data_folder = cir + '分割正解データ削除後/'
suplit_num_file = suplited_correct_data_folder + cir + 'suplit_num'  # 分割数が保存されたファイル
output_node_num_file = suplited_correct_data_folder + '/' + cir + 'suplit_data_num'  # 各モデルの出力ノード数が保存されたファイル
data_correct_rate_file = model_folder + cir + 'data_and_correct_rate' # ノード数，データ数，正解率　⇒　直近の記録
epoch_num_after_added_file = suplited_delited_correct_data_folder + cir + 'epoch_num_after_added'  # 学習回数を追加した後のエポック数を保存するファイル
model_size_file = cir + 'モデルメモリ量/' + cir + '_model_memory_size'


# モデルの分割数を取得
with open(suplit_num_file, 'r') as f:
    num_models = int(f.readline())

print("モデル数:", num_models)

# 入力ノード数を取得　⇒　全モデルで同じ
with open(input_data_file + '0', 'r') as f:
    line = [f.readline().strip().replace("\n", "")]  # 先頭行を読み込む
    input_node_num = len(line[0].strip())

print("入力ノード数:", input_node_num)

# 出力ノード数を取得　⇒　モデルによって異なる可能性がある
with open(output_node_num_file, 'r') as f:
    output_node_num = [int(line.strip().replace("\n", "")) for line in f.readlines()]

print("各モデルの出力ノード数:", output_node_num)
print("len(output_node_num)：", len(output_node_num))


# 全モデルの中間ノード数を取得
with open(data_correct_rate_file, 'r') as f:
    line = f.readlines()  # 先頭2行を読み込む
    top_2_lines = line[:2]
    learning_data = line[2:]  # 3行目以降を読み込む

learning_data_temp = [line.strip().replace("\n", "") for line in learning_data]
middle_node_data_temp = [line.split(",")[1].strip() for line in learning_data_temp if line.strip() != ""]  # 正解率のデータを抽出
middle_node_data = [[0 for _ in range(2)] for _ in range(len(middle_node_data_temp))]  # 初期化
for i in range(len(middle_node_data_temp)):
    middle_node_data[i][0] = int(middle_node_data_temp[i].split("-")[0])
    middle_node_data[i][1] = int(middle_node_data_temp[i].split("-")[1])

with open(epoch_num_after_added_file, 'r') as f:
    lines = f.readlines()
    epochs = [int(line.strip()) for line in lines]

node_group_count = [0] * 9
for i in range(len(epochs)):
    if middle_node_data[i][0] == 25:
        node_group_count[0] += 1
    elif middle_node_data[i][0] <= 50:
        node_group_count[1] += 1
    elif middle_node_data[i][0] <= 75:
        node_group_count[2] += 1
    elif middle_node_data[i][0] <= 100:
        node_group_count[3] += 1
    elif middle_node_data[i][0] <= 125:
        node_group_count[4] += 1
    elif middle_node_data[i][0] <= 150:
        node_group_count[5] += 1
    elif middle_node_data[i][0] <= 175:
        node_group_count[6] += 1
    elif middle_node_data[i][0] <= 200:
        node_group_count[7] += 1
    else:
        node_group_count[8] += 1

epoch_group_count = [0] * 9
for i in range(len(epochs)):
    if epochs[i] <= 1200:
        epoch_group_count[0] += 1
    elif epochs[i] <= 1400:
        epoch_group_count[1] += 1
    elif epochs[i] <= 1600:
        epoch_group_count[2] += 1
    elif epochs[i] <= 1800:
        epoch_group_count[3] += 1
    elif epochs[i] <= 2000:
        epoch_group_count[4] += 1
    elif epochs[i] <= 2200:
        epoch_group_count[5] += 1
    elif epochs[i] <= 2400:
        epoch_group_count[6] += 1
    elif epochs[i] <= 2600:
        epoch_group_count[7] += 1
    else:
        epoch_group_count[8] += 1

print("グループ1（中間層ノード数25,50）のモデル数:", node_group_count[0])
print("グループ2（中間層ノード数50,75）のモデル数:", node_group_count[1])
print("グループ3（中間層ノード数75,100）のモデル数:", node_group_count[2])
print("グループ4（中間層ノード数100,125）のモデル数:", node_group_count[3])
print("グループ5（中間層ノード数125,150）のモデル数:", node_group_count[4])
print("グループ6（中間層ノード数150,175）のモデル数:", node_group_count[5])
print("グループ7（中間層ノード数175,200）のモデル数:", node_group_count[6])
print("グループ8（中間層ノード数200,225）のモデル数:", node_group_count[7])
print("グループ9（中間層ノード数225,250）のモデル数:", node_group_count[8], "\n")

print("グループ1（エポック数1200）のモデル数:", epoch_group_count[0])
print("グループ2（エポック数1400）のモデル数:", epoch_group_count[1])
print("グループ3（エポック数1600）のモデル数:", epoch_group_count[2])
print("グループ4（エポック数1800）のモデル数:", epoch_group_count[3])
print("グループ5（エポック数2000）のモデル数:", epoch_group_count[4])
print("グループ6（エポック数2200）のモデル数:", epoch_group_count[5])
print("グループ7（エポック数2400）のモデル数:", epoch_group_count[6])
print("グループ8（エポック数2600）のモデル数:", epoch_group_count[7])
print("グループ9（エポック数2601以上）のモデル数:", epoch_group_count[8])

print("len(middle_node_data):", len(middle_node_data))

print("num_models:", num_models)
print(len(middle_node_data), "個のモデルの中間ノード数データを取得しました。")
print("len(middle_node_data)：", len(middle_node_data))
print("len(output_node_num)：", len(output_node_num))

# 各モデルの中間ノード数を計算
model_size = 0
for i in range(num_models):
    model_size += 4*((input_node_num+1)*middle_node_data[i][0] + (middle_node_data[i][0]+1)*middle_node_data[i][1] + (middle_node_data[i][1]+1)*output_node_num[i])

print("モデルの総メモリ量(バイト):", model_size/2, "バイト")
print("モデルの総メモリ量(メガバイト):", model_size/2/1024/1024, "MB")  # 1MB = 1024*1024バイト 2で割っているのはTFlite形式でモデルを保存する際に32ビットから16ビットに削減されメモリ量が半分になるため

import os
import datetime as dt
import time

datetime = dt.datetime.now()
os.makedirs(cir + 'モデルメモリ量', exist_ok=True)
with open(model_size_file, 'a') as f:
    f.write('\n')
    f.write(str(datetime) + '\n')
    f.write("実行プログラム：model_size_cal.ipynb\n")
    f.write("モデル分割数：" + str(num_models) + "\n")
    f.write("各モデルの出力ノード数:" + str(output_node_num) + "\n")
    f.write("モデルの総メモリ量(バイト):" + str(model_size/2) + "バイト\n")
    f.write("モデルの総メモリ量(メガバイト):" + str(model_size/2/1024/1024) + "MB\n")
```
