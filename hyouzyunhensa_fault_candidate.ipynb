```python
#グローバル変数
cir = 's15850'  # 対象回路

st_fault_file = cir + 'diagnosis_st_data/' + 'correct_fault_candidate'  # 故障候補数を保存しているファイル
br_fault_file = cir + 'diagnosis_br_data/' + 'correct_fault_candidate'  # 故障候補数を保存しているファイル


def main():
    with open(st_fault_file, 'r') as f:
        lines = f.readlines()
    
    st_fault_candidate =[]
    st_fault_candidate_sum = 0
    st_fault_sum = 0 
    for i in range(len(lines)):
        # print(lines[i].strip().split("：")[0])
        if lines[i].strip().split("：")[0] == "故障候補数（故障の種類ごと）":
            st_fault_candidate.append(int(lines[i].strip().split("：")[1]))
            st_fault_candidate_sum += int(lines[i].strip().split("：")[1])
            st_fault_sum += 1
    

    with open(br_fault_file, 'r') as f:
        lines = f.readlines()
    
    br_fault_candidate =[]
    br_fault_candidate_sum = 0
    br_fault_sum = 0 
    for i in range(len(lines)):
        if lines[i].strip().split("：")[0] == "故障候補数（故障の種類ごと）":
            br_fault_candidate.append(int(lines[i].strip().split("：")[1]))
            br_fault_candidate_sum += int(lines[i].strip().split("：")[1])
            br_fault_sum += 1
    

    st_variance = 0.0
    st_avg = sum(st_fault_candidate)/len(st_fault_candidate)
    for i in range(len(st_fault_candidate)):
        st_variance += (st_fault_candidate[i] - st_avg) ** 2

    hyouzyunhensa = (st_variance / len(st_fault_candidate)) ** 0.5
    print("縮退故障候補の平均：", st_avg)
    print("縮退故障候補の最大値：", max(st_fault_candidate))
    print("縮退故障候補の最小値：", min(st_fault_candidate))
    print("縮退故障候補の標準偏差:", hyouzyunhensa)


    br_variance = 0.0
    br_avg = sum(br_fault_candidate)/len(br_fault_candidate)
    for i in range(len(br_fault_candidate)):
        br_variance += (br_fault_candidate[i] - br_avg) ** 2

    hyouzyunhensa = (br_variance / len(br_fault_candidate)) ** 0.5
    print("ブリッジ故障候補の平均：", br_avg)
    print("ブリッジ故障候補の最大値：", max(br_fault_candidate))
    print("ブリッジ故障候補の最小値：", min(br_fault_candidate))
    print("ブリッジ故障候補の標準偏差:", hyouzyunhensa)
        


if __name__ == "__main__":
    main()
```
